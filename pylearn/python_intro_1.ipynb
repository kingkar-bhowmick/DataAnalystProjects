## Funktionen und dynamische Bindung

Funktionen können mit <code>def()</code> definiert werden. Festlegung der Struktur wieder durch **Einrückung**!

* Es werden keine Typen festgelegt

```python
def zeige(wert):
    print(wert)
```

* Der Typ wird automatisch aus dem Kontext abgeleitet

```python
x = 3
zeige(x)
```

* Ein Typ kann jederzeit geändert werden

```python
x = "String"
zeige(x)
```

## Objektorientierung

Definition einer Klasse:

```python
class Klasse:
    # Klassenvariable
    anzahl = 0
    
    # Der Konstruktor
    def __init__(self, name):
        self.name = name
        Klasse.anzahl += 1
    
    # Eine Instanzmethode
    def methode1(self, parameter):
        if parameter == 1:
            print("%s: Eins" % self.name)
        elif parameter == 2:
            print("%s: Zwei" % self.name)
        else:
            print("Anzahl: %i" % Klasse.anzahl)
    
    # Eine Klassenmethode
    def methode2(parameter):
        print(parameter)
```

Erstellung eines Objekts und Aufruf einer Instanzmethode:

```python
objekt1 = Klasse("Objekt 1")
objekt1.methode1(1)
objekt1.methode1(2)
objekt1.methode1(3)
```

```python
objekt2 = Klasse("Objekt 2")
objekt2.methode1(1)
objekt2.methode1(5)
```

Aufruf einer Klassenmethode:

```python
Klasse.methode2("Klassenmethode")
```
