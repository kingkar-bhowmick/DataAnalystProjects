## 2. Everything is an object!
### 2.a Lists

```python
# Most common datatypes in Python are lists, tuples and dictionaries (sets are important too but not described here)
a = [1, 2, 3, 4, 5, 6, 7, 10, "9"] # A list is a FILO stack comparable to arrays in other languages
print(a, "has", len(a), "elements")
print(type(a))
b = a.pop() # removes the last entry and returns it
print(b)
c = a.pop(0) # removes the entry at the given index and returns it
print(c)
print(a, "has", len(a), "elements")
# If we do not want to remove the entries, we can instead use indexing
print(a[1]) # Index starts from 0 so this returns the second entry in a
print(a[-1], a[6]) # Negative index is equivalent to len(a) - idx so this returns the last element
print(a[:])
print(a[:2], a[2:], a[0:4], a[::2]) # A slice has none or more of start:stop:step, where stop idex is not included
print(a[:-2], a[-2:], a[-4:-1], a[-1:-4:-2]) # All entries of a slice can be negative
```

```python
# lists can be added, extended, and appended (to)
print(a)
c = [4, 5, 6]
print(c)
b = a + c # Add two list
print(b)
d = b + a
b.extend(a) # extend b with the list a
print(b)
print(d)
b.append(9) # append 8 to b
print(b)
b[0] = 1 # change the value of 0-th entry to 1
print(b)
```

```python
# Let us look at what we mean by mutability 
a = [1, 2, 3] # a list is mutable
b = a # so b is pointing to a instead of copying it
b.append(4)
print(a, b)
print(a is b) # is compares equality instead of equivalency, if a is b, then they both point to the same object
# Therefore, mutable objects should be handled with care to avoid bugs.
# atomic types (floats, ints, strings, booleans) are immutable. if a list only has immutable objects
# Following would yield more expected results
c = a.copy() # we are calling the copy method of a, which is a list
c.append(5)
print(a, c, a is c)
d = a[:]
d.append(5)
print(a, d, a is d)
```

### 2.c Dictionaries

```python
# The last data structure we will look into is a dictionary, which is a hash table in python. 
# It consists of key value pairs where the order is not guaranteed to be preserved
# Dictionaries are mutable.
a = {"my": 0, "first": 1, "dict": 2}
print(a, type(a))
print(a.keys(), a.values()) # We can access keys and values individually
print(a.items()) # But we often acces both at the same time if we want to iterate through them
print(a["my"], a["first"], a["dict"]) # We can also access the entries directly using their keys.
a["my"] = "a"
a["add"] = 3
print(a)
b = {"my": 1, "second": 2}
a.update(b)
print(a)
```

### 2.d Functions

```python
# Other than classes, functions are a handy way to write units of work in python. 
# A function is defined using def keyword
def fun():
    """prints well played"""
    base = ["Well", "played"]
    print(" | ".join(base))

# Now instead of possibly writing a lot of code, we can just call the function
fun()
fun()
```

```python
# Similar the methods, functions acceps arguments and keyword arguments
# It is almost alway meaningfull to use parameters instead of hardcoding things
def have_fun(name, msg=("Well", "Played"), sep=" "):  # Using immutable defaults is important
    base = msg + (name, ) # (name, ) defines a tuple but (name) does not!
    print(sep.join(base))
have_fun("Can")
have_fun("Jule", msg=("Good", "Luck"))
```

### 2.e Classes and Objects

```python
# But wait, we haven't talked about objects yet. A class in python is defined using the class keyword
class A:
    name = "Can" #This is a class attribute

b = A() # We now instatiate a class A and assign it to b
print(A, b) # Thus A is a class and b is an object
print(A.name, b.name) # Both classes and instances can access the class attributes
# The following is a bad idea..
class B:
    bag = []
# Because this is not intuitive
c = B()
d = B()
print(c, d, c is d)
c.bag.append("laptop")
print(c.bag, d.bag)
```

```python
# For this and other reasons, using the __init__ function is often better, which is called during instance creation
class C:
    def __init__(self): # self refers to the instance that is being created.
        self.bag = []  # we set the bag property of self to an empty list
# C() 
#     => c.__new__=> self
#     => c.__init__(self)
# => self
a = C()
b = C()
a.bag.append("laptop")
print(a.bag, b.bag) # This is what we expected to have as an output in the previous cell
# Don't forget that you cannot access the attributes from a class, which are created at __init__
print(C().bag)
print(C.bag)
```

```python
# We can also parametrize the __init__
class B:
    def __init__(self, contents): # contents is a positional argument
        # Positional arguments are accessed by the order index of the inputs
        self.bag = contents.copy()
a = B(["laptop"])
print(a.bag)
# Pro tip: you can change attributes of any object on the fly using assignment (see monkey patching)
a.bag = ["beer", "more beer", "keys"]
a.lost = ["laptop"]
print(a.bag, a.lost)
```

```python
a = [1, 2, 3]
b = (1, 2, 3)
c = {1: "0", "1": 2, "2": 3}
print(1 in a, 4 in a)
print(1 in b, 4 in b)
print(1 in c, 3 in c)
```

```python
# Nonetheless, it is much better to write methods to change the object state
class B:
    def __init__(self, contents=None): # content is a keyword argument
        # contrary to positional arguments, keyword arguments allow setting default values
        # keyword arguments are either accessed by key or by order index depending on the call
        # Pro tip: Always prefer keys so B(contents=[]) instead of B([])
        if contents is None:
            contents = []
        self.bag = contents.copy()
        self.lost = []

    def replace_contents(self, new_contents, lost_contents=None):
        self.bag = new_contents.copy()
        if lost_contents is not None: # one of the few cases, where is comparison should be preferred
            self.lost = lost_contents.copy()
    
    def lose(self, item):
        if item in self.bag: # Check if item is in the bag
            self.bag.remove(item)
            self.lost.append(item)
        else:
            print(item, "was not in bag.")

a = B()
a.replace_contents(["laptop"])
print(a.bag, a.lost)
a.lose("laptop")
print(a.bag, a.lost)
a.lose("beer")
a.replace_contents(["beer", "more beer", "keys"])
print(a.bag, a.lost)
a.replace_contents(["beer"], lost_contents=["more beer", "keys", "laptop"])
print(a.bag, a.lost)
```

```python
# Let us import some useful libraries for mathematical analysis and visualisation
import numpy as np  # Useful for anything math (on CPU)
import pandas as pd # Useful for data operations on "table-ish" data
import matplotlib.pyplot as plt # Useful for plots
import seaborn as sns # Also useful for plots
import plotly.express as px # Also useful for plots???

# Pro tip: It is a good practice to have all your imports at the beginning of your files
```

```python
# Create some standard normal data
np.random.seed(0) # Ignore for now
X = np.random.randn(10, 4) # This returns a numpy array
print(X) # A numpy array is similar to a list but may have a higher rank
print(X.shape)
print(X.ndim)
```

```python
y = np.sum(X, axis=1) # Sum the entries of X along the first dimension
print(y) # A numpy array is similar to a list but may have a higher rank
print(y.shape)
print(y.ndim)
```

```python
# many common operations are already implemented in numpy (though searching may be required to find them)
C = np.corrcoef(X, rowvar=False) # What does rowvar=False mean? Read the docs! (RTFM)
print(C)
```

```python
# Similar to lists, we can append array of same rank
y = y.reshape((-1, 1)) # reshape y, such that it has 1 column and as many rows as necessary to match the shape
print(y.shape)
D = np.append(X, y, axis=1) # Append y to X along columns
print(D)
print(D.shape)
print(D.ndim)
C2 = np.corrcoef(D, rowvar=False) # What does rowvar=False mean? Read the docs! (RTFM)
print(C2)
```

```python
# So if numpy can do so much, what does pandas do? It brings order to chaos (among other useful things)
input_names = ["X1", "X2", "X3", "X4"]
df = pd.DataFrame(D, columns=input_names + ["y"])
print(df)
```

```python
# most array operations are also defined for pandas dataframes
print(np.mean(D, axis=0))
print(np.mean(df, axis=0))
# Probably the most useful thing about numpy is reading table-formed data such as .csv, .xls etc.
# Checkout the docs for more detail.
```

```python
# Now let us visualise some data to finish this notebook
fig, ax = plt.subplots(5, 5, figsize=(12, 12)) # Creates a plot with a 5x5 grid of subplots
for row in range(5):
    for col in range(5):
        cur_ax = ax[row, col]
        if row > col:
            cur_ax.scatter(D[:, col], D[:, row])
        elif row == col:
            cur_ax.hist(D[:, row])
        else:
            corr = C2[col, row]
            cur_ax.text(0.5, 0.5, f"{corr:.2f}", fontsize=16, horizontalalignment="center",
                        verticalalignment="center")
            cur_ax.set_yticks([])
            col = (1, 1 - abs(corr), 1 - abs(corr))
            cur_ax.set_facecolor(col)
            cur_ax.set_xticks([])

_ = plt.figure() # We can also plot faster using plt directly
_ = plt.imshow(C2) # Useful for experimenting, too limiting for usage in code
# Checkout https://matplotlib.org/stable/gallery/index.html for more
```

```python
# So if matplotlib can do so much, what can seaborn do? It also uses matplotlib but has useful
# built ins plots as extension to matplotlib. It works with dataframes
g = sns.pairplot(df) # Much simpler, although it is missing the correlations
```

```python
# we can add the correlation manually using map
def mid_point(x):
    return 0.5 * (x.min() + x.max())

def add_corr(x, y, ax=None, **kws):
    """Plot the correlation coefficient in the top left hand corner of a plot."""
    corr = np.corrcoef(np.c_[x, y], rowvar=False)[0, 1]
    cur_ax = ax or plt.gca()
    mid_x = mid_point(x)
    mid_y = mid_point(y)
    cur_ax.text(mid_x, mid_y, f"{corr:.2f}", fontsize=16, horizontalalignment="center",
                verticalalignment="center")
    cur_ax.set_yticks([])
    col = (1, 1 - abs(corr), 1 - abs(corr))
    cur_ax.set_facecolor(col)
    cur_ax.set_xticks([])

g = sns.pairplot(df) # Much simpler, although it is missing the correlations
_ = g.map_upper(add_corr)
# Check out https://seaborn.pydata.org/examples/index.html for more
```
